#**MultiEURLEX**

MultiEURLEX est un dataset comprenant **65 000 lois** de l'UE disponibles dans **23 langues** officielles. Chaque loi est annotée avec des concepts EUROVOC (EUROVOC est un outiL pour la gestion de l'information au sein des institutions de l'UE), permettant une classification uniforme et facilitant l'accès aux documents dans plusieurs langues., associant des identifiants à des descripteurs tels que [60, produits agroalimentaires] et [1115, fruit], disponibles dans toutes les langues.

 MultiEURLEX peut être utilisé pour la **classification de sujets juridiques**, une tâche multi-labels où les documents doivent se voir attribuer des concepts EUROVOC reflétant leurs sujets. Cependant, MultiEURLEX prend en charge les étiquettes de trois granularités différentes (niveaux EUROVOC). Il permet également d'étudier des scénarios de transfert interlingue, incluant un à plusieurs, et plusieurs à plusieurs langues.

La plateforme permet aux utilisateurs de rechercher, consulter et télécharger ces textes légaux, assurant ainsi la fiabilité des informations juridiques et facilitant l'accès à la législation en vigueur en France.

#**FCD (Décisions des Tribunaux français)**



Le jeu de données contient des **décisions des tribunaux français**, notamment de la Cour d'appel et de la Haute Cour. Chaque décision est structurée en métadonnées détaillant le tribunal, le numéro de procès et la date, suivies par les parties impliquées. La composition de la cour est également spécifiée, incluant le président du tribunal et le greffier. Les faits de l'affaire sont présentés, suivis des moyens et principaux arguments avancés par les parties. Les motifs du jugement sont détaillés, comprenant les **arguments utilisés par le tribunal pour parvenir à sa décision finale**. Enfin, le dispositif du jugement précise la décision finale du tribunal. Les données sont en **français**, assurant ainsi une grande précision et pertinence pour les utilisateurs francophones.

Disponible [ici](https://datasets.doctrine.fr/)

#**MultiLegalSBD**


MultiLegalSBD est un dataset **multilingue** destiné à la détection des frontières de phrases dans des documents juridiques. Il comprend des textes juridiques annotés pour indiquer précisément où commencent et finissent les phrases. Utilisé pour entraîner des modèles de traitement du langage naturel (NLP), ce dataset améliore la qualité des applications NLP juridiques telles que le **résumé automatique** et l'**extraction d'informations**.

Disponible sur [Hugging Face](https://huggingface.co/datasets/rcds/MultiLegalSBD)
